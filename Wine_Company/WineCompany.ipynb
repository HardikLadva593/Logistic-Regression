```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X = wine.drop(['Unnamed: 0','quality'], axis=1)
X.head()
```

```python
Y = wine['quality']
Y.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3,random_state = 10)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

LR = LogisticRegression()
OneVsRes= OneVsRestClassifier(LR)
```

```python

# Fitting the model with training data
OneVsRes.fit(X_train, y_train)
prediction_oneVsRest = OneVsRes.predict(X_test_scaled )
X_test['prob_oneVsrest'] = OneVsRes.predict_proba(X_test_scaled ).tolist()
X_test['prob_oneVsrest_a'] = OneVsRes.predict_proba(X_test_scaled )[:,0].tolist()
X_test['prob_oneVsrest_b'] = OneVsRes.predict_proba(X_test_scaled )[:,1].tolist()
X_test['prob_oneVsrest_c'] = OneVsRes.predict_proba(X_test_scaled )[:,6].tolist()
results_interpretation = OneVsRes.coef_.shape
```
